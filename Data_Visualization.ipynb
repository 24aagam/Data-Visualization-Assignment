# 4. Using the given x and y data, generate a 3D surface plot to visualize the function image.png

```python
import numpy as np
import pandas as pd
```

```python

# Given x, y, and z data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create the 3D surface plot
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])
fig.update_layout(scene=dict(zaxis_title="Z", xaxis_title="X", yaxis_title="Y"),
                  title="3D Surface Plot of $z = \sin(\sqrt{x^2 + y^2})$")

# Show the plot
fig.show()
```

# 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256

# Generate random data for x, y, sizes, and colors
n_points = 100
x_values = np.random.rand(n_points)
y_values = np.random.rand(n_points)
sizes = np.random.randint(5, 20, size=n_points)  # Random sizes
colors = np.random.randint(0, 256, size=n_points)  # Random color indices

# Create a DataFrame to hold the data
df = pd.DataFrame({
    'x': x_values,
    'y': y_values,
    'sizes': sizes,
    'colors': colors
})

# Create a Bokeh ColumnDataSource
source = ColumnDataSource(df)

# Define a color mapper based on the 'colors' column
color_mapper = linear_cmap(field_name='colors', palette=Viridis256, low=min(colors), high=max(colors))

# Create a Bokeh figure
p = figure(title="Random Scatter Plot", x_axis_label="X", y_axis_label="Y")

# Add the scatter plot with custom sizes and colors
p.scatter(x='x', y='y', size='sizes', color=color_mapper, source=source, legend_field='colors')

# Add color bar
color_bar = p.legend[0]
color_bar.title = 'Colors'
color_bar.label_standoff = 12

# Show the plot
show(p)
```
